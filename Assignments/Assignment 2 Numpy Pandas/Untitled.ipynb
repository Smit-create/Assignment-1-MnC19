```python
import numpy as np
a = np.arange(120).reshape((10,3,4))
print(a)
b = np.arange(120)[::-1].reshape((3,4,10))
print(b)
np.dot(a, b)[0,0,0,0]
```

```python
Z = np.ones((10,10))
Z[1:-1,1:-1] = 0
print(Z)
```

```python
Z = np.ones((5,5))
Z = np.pad(Z, pad_width=1, mode='constant', constant_values=0)
print(Z)

# Using fancy indexing
Z[:, [0, -1]] = 0
Z[[0, -1], :] = 0
print(Z)
```
